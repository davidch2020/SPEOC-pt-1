```python
LoanOffice = pd.read_csv('../../Data/Pre1790/cleaned/loan_office_certificates_9_states_cleaned.csv', index_col=0)
LoanOffice = LoanOffice[['First Name 1 ', 'Last Name 1 ', 'Face Value']]
LoanOffice['Name'] = LoanOffice['First Name 1 '].apply(lambda x: "" if pd.isnull(x) else x) + ' ' + LoanOffice['Last Name 1 '].apply(lambda x: "" if pd.isnull(x) else x)
LoanOffice.drop(['First Name 1 ', 'Last Name 1 '], inplace=True, axis=1)
LoanOffice.rename(columns = {'Face Value' : 'Dollars'}, inplace = True)
```

```python
Marine = pd.read_csv('../../Data/Pre1790/cleaned/Marine_Liquidated_Debt_Certificates_cleaned.csv', index_col=0)
Marine = Marine[['First name', 'Last name', 'Dollars']]
Marine['Name'] = Marine['First name'].apply(lambda x: "" if pd.isnull(x) else x) + ' ' + Marine['Last name'].apply(lambda x: "" if pd.isnull(x) else x)
Marine.drop(['First name', 'Last name'], inplace=True, axis=1)
```

```python
Pierce = pd.read_csv('../../Data/Pre1790/cleaned/Pierce_Certs_cleaned_2021.csv', index_col=0)
Pierce = Pierce[['First', 'Last', 'Value']]
Pierce['Name'] = Pierce['First'].apply(lambda x: "" if pd.isnull(x) else x) + ' ' + Pierce['Last'].apply(lambda x: "" if pd.isnull(x) else x)
Pierce.drop(['First', 'Last'], inplace=True, axis=1)
Pierce.rename(columns = {'Value' : 'Dollars'}, inplace = True)
```

```python
def Search(Person):
    PACopy = PA
    PACopy['Ratio'] = PACopy['Name'].apply(lambda Name: fuzz.ratio(str(Name), Person))
    
    print('PA Data')
    print('-------')
    print(PACopy.loc[PACopy['Ratio'] >= 80])
    print('\n')
    
    LOCopy = LoanOffice
    LOCopy['Ratio'] = LOCopy['Name'].apply(lambda Name: fuzz.ratio(str(Name), Person))
    
    print('Loan Office')
    print('-----------')
    print(LOCopy.loc[LOCopy['Ratio'] >= 80])
    print('\n')
    
    MarineCopy = Marine
    MarineCopy['Ratio'] = MarineCopy['Name'].apply(lambda Name: fuzz.ratio(str(Name), Person))
    
    print('Marine')
    print('------')
    print(MarineCopy.loc[MarineCopy['Ratio'] >= 80])
    print('\n')
    
    PierceCopy = Pierce
    PierceCopy['Ratio'] = PierceCopy['Name'].apply(lambda Name: fuzz.ratio(str(Name), Person))
    
    print('Pierce')
    print('------')
    print(PierceCopy.loc[PierceCopy['Ratio'] >= 80])
    print('\n')
    
    return PACopy, LOCopy, MarineCopy, PierceCopy
```

```python
TableIndexes = {'PA': 0, 'LO': 1, 'Marine': 2, 'Pierce': 3}
```

```python
def FindTotal(Data, Exceptions):
    Sum = 0
    for Table in Data:
        Table = Table.loc[Table['Ratio'] == 100]
        Sum += Table['Dollars'].sum()  
            
    if Exceptions is not None:
        for Exception in Exceptions.keys():
            for Index in Exceptions[Exception]:
                Sum += Data[TableIndexes[Exception]].loc[Index, 'Dollars']
    
    return Sum
```
