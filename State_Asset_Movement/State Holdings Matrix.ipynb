```python
import numpy as np
import pandas as pd
```

```python
geog_order = ['NH', 'VT', 'MA', 'RI', 'CT', 'NY', 'PA', 'NJ', 'DE', 'MD', 'VA', 'NC', 'SC', 'GA']
multi_CD = pd.DataFrame(columns=geog_order)
```

## Continental Debt by 6p Dollar

```python
#uses only sum of 6 percent dollar
def sum_by_state(df, multi):
    df.columns = ['state', '6p_Dollar']
    df.loc[df[df['state'].apply(lambda x: pd.isnull(x))].index, 'state'] = 'unspecified'
    df_statewise = df.groupby('state')
    #df_count = df_statewise.count()
    df_sum = df_statewise.agg('sum').transpose()
    return multi.merge(df_sum, how='outer')
```

```python
NH_CD = pd.read_excel("../Data/Post1790/NH/T652_R6_New_Hampshire_CD.xlsx", header = 10, usecols = 'L, N')
multi_CD = sum_by_state(NH_CD, multi_CD)

RI_CD = pd.read_excel("../Data/Post1790/RI/T653_Rhode_Island_CD.xlsx", header = 11, usecols = 'J, L')
multi_CD = sum_by_state(RI_CD, multi_CD)

CT_CD = pd.read_excel("../Data/Post1790/CT/CT_post1790_CD_ledger.xlsx", header = 13, usecols = 'K, N')
multi_CD = sum_by_state(CT_CD, multi_CD)

NY_CD = pd.read_excel("../Data/Post1790/NY/NY_1790_CD.xlsx", header = 11, usecols = 'M, P')
NY_CD.columns = ['6p_Dollar', 'state']
NY_CD.loc[NY_CD[NY_CD['state'].apply(lambda x: pd.isnull(x))].index, 'state'] = 'unspecified'
NY_CD_statewise = NY_CD.groupby('state')
NY_CD_sum = NY_CD_statewise.agg('sum').transpose()
multi_CD = multi_CD.merge(NY_CD_sum, how='outer')

PA_CD = pd.read_excel("../Data/Post1790/PA/PA_post1790_CD.xlsx", header = 11, usecols = 'J, L')
multi_CD = sum_by_state(PA_CD, multi_CD)

MD_CD = pd.read_excel("../Data/Post1790/MD/MD_post1790_CD.xlsx", header = 11, usecols = 'J, L')
multi_CD = sum_by_state(MD_CD, multi_CD)

VA_CD = pd.read_excel("../Data/Post1790/VA/VA_CD.xlsx", header = 11, usecols = 'K, M')
VA_CD.columns = ['6p_Dollar', 'state']
VA_CD.loc[VA_CD[VA_CD['state'].apply(lambda x: pd.isnull(x))].index, 'state'] = 'unspecified'
VA_CD_statewise = VA_CD.groupby('state')
VA_CD_sum = VA_CD_statewise.agg('sum').transpose()
multi_CD = multi_CD.merge(VA_CD_sum, how='outer')

NC_CD = pd.read_excel("../Data/Post1790/NC/T695_R4_NC_CD.xlsx", header = 11, usecols = 'M, W')
multi_CD = sum_by_state(NC_CD, multi_CD)

SC_CD = pd.read_excel("../Data/Post1790/SC/Post_1790_South_Carolina_CD.xlsx", header = 11, usecols = 'G, M')
multi_CD = sum_by_state(SC_CD, multi_CD)

GA_CD = pd.read_excel("../Data/Post1790/GA/T694_GA_Loan_Office_CD.xlsx", header = 10, usecols = 'T, Z')
multi_CD = sum_by_state(GA_CD, multi_CD)
```

```python
multi_CD
```

```python
merge_order = pd.Series(['NH', 'RI', 'CT', 'NY', 'PA', 'MD', 'VA', 'NC', 'SC', 'GA'])
column_order = geog_order + ['unspecified', 'BVI', 'BM', 'FR', 'GB', 'US', 'VI']
multi_CD.set_index(merge_order, inplace=True)
multi_CD = multi_CD.reindex(index=geog_order)
multi_CD = multi_CD.reindex(columns=column_order)
```

```python
multi_CD
```

```python
def highlight_diag(df):
    a = np.full(df.shape, '', dtype='<U24')
    np.fill_diagonal(a, 'background-color: yellow')
    return pd.DataFrame(a, index=df.index, columns=df.columns)

s = multi_CD.style.format(na_rep='', precision=0)
s = s.apply(highlight_diag, axis=None)
s
```

## Continental Debt by percent

```python
state_total = pd.Series(multi_CD[column_order].sum(axis = 1))
grand_total = state_total.sum()
multi_CD_percent = multi_CD.div(state_total, axis='index') * 100
state_percent = state_total / grand_total * 100
multi_CD_percent['total'] = state_percent
```

```python
multi_CD_percent
```

```python
s2 = multi_CD_percent.style.format(na_rep='', precision=2).hide_columns(['BVI', 'BM', 'FR', 'GB', 'US', 'VI'])
s2 = s2.apply(highlight_diag, axis=None)
s2
```
